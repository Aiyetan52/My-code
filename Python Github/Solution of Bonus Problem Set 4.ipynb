## Question 1
(10 pts) Suppose that the values for the parameters in problem 1a are m=100, p=2 and q=1, Solve for optimal demand levels for both goods and the highest possible utility level.

```python
import sympy as sp
import sympy 
from sympy import symbols 
from sympy import *
sympy.init_printing()
#P=2 # price of F
#q=1 #Price of Y
#I=12 # Income
p,q,m,x,y,=sympy.symbols(('p','q','m','x','y'))
x,y = sp.var('x,y',real=True);#Define the choice variables 
U = ((1/2)*log(x))+((1/3)*log(y)) #Utility function 
g = p*x+ q*y - m #Budget constraint i.e g=0
lam = sp.symbols('lambda', real = True) # the exchange rate variable of lagrangian multiplier
L = U - lam* g # lagrangian function
gradL = [sp.diff(L,c) for c in [x,y]] # Gradient of Lagrangian w.r.t. (x,y)
KKT = gradL + [g]# Determine the Karush-Kuhn-Tucker(KKT) condition 
Sol= sp.solve(KKT, [x, y, lam], dict=True) #minimize or maximize the U given g=0 by solving KKT conditions over F,Y Lambda
Sol
```

```python
import sympy as sp
import sympy 
sympy.init_printing()
p,q,U,x,y,=sympy.symbols(('p','q','U','x','y'))
x, y = sp.var('x,y',real=True);#Define the choice variables 
h= ((1/2)*log(x))+((1/3)*log(y))-U #Demand Equation 
g=p*x+ q*y #Cost function i.e g=0
lam = sp.symbols('lambda', real = True) # the exchange rate variable of lagrangian multiplier
L = g - lam* h # lagrangian function
gradL = [sp.diff(L,c) for c in [x,y]] # Gradient of Lagrangian w.r.t. (x,y)
KKT = gradL +[h] # Determine the Karush-Kuhn-Tucker(KKT) condition 
Sol= solve(KKT, [x, y, lam], dict=True) #minimize or maximize the U given g=0 by solving KKT conditions over F,Y Lambda
Sol # Note: ignore solution with imaginary/complex numbers
```
